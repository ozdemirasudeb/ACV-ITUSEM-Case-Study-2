Bu notebook'ta elektrik tüketim verileri yüklenmiş, veri setlerinin yapısı incelenmiş ve ilk keşif analizleri gerçekleştirilmiştir.

Bu kapsamda:

```python
import pandas as pd
import numpy as np

file_path = "/content/elektrik_veri_hashed.xlsx"

xls = pd.ExcelFile(file_path)

print("Sayfa isimleri:")
print(xls.sheet_names)
```

```python
df_hamamozu = pd.read_excel(
    file_path,
    sheet_name="Tahakkuk"
)

df_hamamozu.shape
```

```python
df_gumushacikoy = pd.read_excel(
    file_path,
    sheet_name="Tahakkuk 1"
)

df_gumushacikoy.shape
```

```python
df_goynucek = pd.read_excel(
    file_path,
    sheet_name="Tahakkuk 2"
)
df_goynucek.shape
```

```python
print("Hamamözü:", df_hamamozu.shape)
print("Gümüşhacıköy:", df_gumushacikoy.shape)
print("Göynücek:", df_goynucek.shape)
```

```python
print("Hamamözü:", df_hamamozu["sozlesme_hesap_no"].nunique())

print("Gümüşhacıköy:", df_gumushacikoy["sozlesme_hesap_no"].nunique())

print("Göynücek:", df_goynucek["sozlesme_hesap_no"].nunique())
```

```python
df_tahakkuk = pd.concat(
    [df_hamamozu,
     df_gumushacikoy,
     df_goynucek],
    ignore_index=True
)

df_tahakkuk.shape
```

```python
print(df_tahakkuk.shape)
```

```python
print("Eksik kWh değeri:", df_tahakkuk["kwh"].isna().sum())
```

```python
print("Negatif kWh sayısı:", (df_tahakkuk["kwh"] < 0).sum())
```

```python
Q1 = df_tahakkuk["kwh"].quantile(0.25)
Q3 = df_tahakkuk["kwh"].quantile(0.75)

IQR = Q3 - Q1

alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

print("Alt sınır:", alt_sinir)
print("Üst sınır:", ust_sinir)
```

```python
aykiri = df_tahakkuk[
    (df_tahakkuk["kwh"] < alt_sinir) |
    (df_tahakkuk["kwh"] > ust_sinir)
]

print("Aykırı değer sayısı:", len(aykiri))
```

```python
aykiri.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.boxplot(df_tahakkuk["kwh"], vert=False)

plt.title("kWh Tüketim Dağılımı")
plt.xlabel("kWh")

plt.show()
```

kwh sütununda eksik değer bulunmamaktadır. 151 adet negatif tüketim kaydı tespit edilmiştir. IQR yöntemi ile normal dağılımın dışında kalan aykırı değerler belirlenmiştir.

```python
hesap_ozet = (
    df_tahakkuk
    .groupby("Hesap Sınıfı")["kwh"]
    .agg(
        Ortalama="mean",
        Medyan="median",
        Std="std",
        Minimum="min",
        Maksimum="max",
        Kayit_Sayisi="count"
    )
    .round(2)
)

hesap_ozet
```

```python
hesap_ozet.sort_values("Ortalama", ascending=False)
```

Elektrik tüketimi hesap sınıfına göre belirgin biçimde değişmektedir. Kurumsal ve üretim amaçlı aboneler en yüksek tüketimi gerçekleştirirken mesken aboneleri en geniş müşteri grubunu oluşturmasına rağmen ortalama tüketimleri daha düşüktür. Bu nedenle tüketim analizlerinde hesap sınıfı önemli bir ayırıcı değişken olarak değerlendirilmelidir.

```python
ilce_ozet = (
    df_tahakkuk
    .groupby("ilce")["kwh"]
    .agg(
        Ortalama="mean",
        Medyan="median",
        Std="std",
        Kayit="count"
    )
    .round(2)
)

ilce_ozet
```

```python
genel_ozet = pd.DataFrame({
    "Toplam Kayıt": [len(df_tahakkuk)],
    "Toplam Sütun": [df_tahakkuk.shape[1]],
    "Benzersiz Müşteri": [df_tahakkuk["sozlesme_hesap_no"].nunique()],
    "İlçe Sayısı": [df_tahakkuk["ilce"].nunique()],
    "Hesap Sınıfı Sayısı": [df_tahakkuk["Hesap Sınıfı"].nunique()],
    "Eksik kWh": [df_tahakkuk["kwh"].isna().sum()],
    "Negatif kWh": [(df_tahakkuk["kwh"] < 0).sum()],
    "Ortalama kWh": [round(df_tahakkuk["kwh"].mean(), 2)],
    "Medyan kWh": [round(df_tahakkuk["kwh"].median(), 2)]
})

display(genel_ozet)
```

Birleştirilen veri seti toplam 1.185.698 kayıttan oluşmaktadır. Veri setinde 3 ilçe ve 37 hesap sınıfı bulunmaktadır. kwh sütununda eksik değer bulunmazken 151 negatif tüketim kaydı tespit edilmiştir. Ortalama tüketim 92,64 kWh, medyan tüketim ise 46,62 kWh olarak hesaplanmıştır.
